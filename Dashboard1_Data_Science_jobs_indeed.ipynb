```python
from bs4 import BeautifulSoup 
import urllib2 
import re 
from time import sleep 
from collections import Counter 
from nltk.corpus import stopwords 
import pandas as pd 
```

```python
def text_cleaner(website):
    try:
        site = urllib2.urlopen(website).read() 
    except: 
        return   
    soup_obj = BeautifulSoup(site) 
    
    for script in soup_obj(["script", "style"]):
        script.extract() 
    
    

    text = soup_obj.get_text() 
    
        
    
    lines = (line.strip() for line in text.splitlines()) 
    
        
        
    chunks = (phrase.strip() for line in lines for phrase in line.split("  ")) 
    
    def chunk_space(chunk):
        chunk_out = chunk + ' ' 
        return chunk_out  
        
    
    text = ''.join(chunk_space(chunk) for chunk in chunks if chunk).encode('utf-8') 
        
        
        
    try:
        text = text.decode('unicode_escape').encode('ascii', 'ignore') 
    except:                                                            
        return                                                         
       
        
    text = re.sub("[^a-zA-Z.+3]"," ", text)  
                                               
        
       
    text = text.lower().split()  
        
        
    stop_words = set(stopwords.words("english")) 
    text = [w for w in text if not w in stop_words]
        
        
        
    text = list(set(text)) 
                            
        
    return text
```

```python
import geocoder
def get_latitude(row):
    location="'"+row['city']+"','"+row['state']
    coordinates = geocoder.google(location)
    return coordinates.lat
def get_longitude(row):
    location="'"+row['city']+"','"+row['state']
    coordinates = geocoder.google(location)
    return coordinates.lng

```

```python
jobs_data_scientist['latitude']=jobs_data_scientist.apply(get_latitude,axis=1)
jobs_data_scientist['longitude']=jobs_data_scientist.apply(get_longitude,axis=1)
```
