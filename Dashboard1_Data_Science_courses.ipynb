```python
#normalise the data and group courses of similar nature or title together 
def course_type(row):
    if 'Master of Science' in row['PROGRAM'] or  'M.S.' in row['PROGRAM']:
        val='Science-Masters'
    elif 'Ph.D.' in row['PROGRAM']:
        val='Doctoral'
    elif 'certificate' in row['PROGRAM']:
        val='Certificate'
    elif 'Arts' in row['PROGRAM']:
        val='Arts- Masters'
    elif 'Business Analytics' in row['PROGRAM']:
        val='Masters in Business Analytics'
    elif 'Technology' in row['PROGRAM']:
        val='Engineering- Masters'
    elif 'Business' in row['PROGRAM'] or 'MBA' in row['PROGRAM']:
        val='Business- Masters'
    else:
        val='General'
    return val
        
```

```python
data_science_data_2['course_type']=data_science_data_2.apply(course_type,axis=1)
```

```python
#normalise the data based ont he department they are affiliated to 
def department_type(row):
        if 'Computer Science' in row['DEPARTMENT'] or 'Computing' in row['DEPARTMENT']:
            val='Computer Science' 
        elif 'Science' in row['DEPARTMENT']:
            val='Science'
        elif 'Statistic' in row['DEPARTMENT']:
            val='Statistics'
        elif 'Business' in row['DEPARTMENT']:
            val='Business'
        elif 'Informatic' in row['DEPARTMENT']:
            val='Informatics'
        else:
            val='Other'
        return val
        
```

```python
data_science_data_2['department_type']=data_science_data_2.apply(department_type,axis=1)
```

# Using the santa clara standards to compare the duration to the number of credits. At Santa Clara we are advised to pursue the course at 3 credits / a month

```python
import re
def credit_time_func(row):
    m = re.search('([0-9].*).*[cC]redit',row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1))
        try:
            return  int(m.group(1))
        except:
            pass
    m = re.search('([0-9].*).*[mM]onth', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1)) * 3
        try:
            return int(m.group(1))*3
        except:
            pass
    m = re.search('([0-9].*).*[uU]nit', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1))
        try:
            return int(m.group(1))
        except:
            pass
    m = re.search('([0-9].*).*[yY]ear', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1)) * 12 * 3
        try:
            return int(m.group(1))*12*3
        except:
            pass
    m = re.search('([0-9].*).*[hH]our', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1))/12
        try:
            return int(m.group(1))/12
        except:
            pass
    m = re.search('([0-9].*).*[cC]ourses', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1))*3
        try:
            return int(m.group(1))*3
        except:
            pass
    m = re.search('([0-9].*).*[cC]EU', row['DURATION'])
    if m:
        s = re.search('([0-9].*)-([0-9].*)', m.group(1))
        if s:
            return int(s.group(1))
        try:
            return int(m.group(1))
        except:
            pass
    else:
        return 0
```

```python
data_science_data_2['credits']=data_science_data_2.apply(credit_time_func,axis=1)
```

```python
credit_mean_value=data_science_data_2['credits'].mean()
credit_std=data_science_data_2['credits'].std()
print credit_mean_value, credit_std
```
